```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import tkinter as tk
```

```python
df = pd.read_csv("Admissions.csv")
df
```

```python
X = df[['PCM', 'Sports', 'Project', 'Warnings']]
y = df['Admitted']
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

## Use a ML algorithm

```python
clf = RandomForestClassifier()
clf.fit(X, y)
```

```python
# tkinter GUI
root = tk.Tk()

canvas1 = tk.Canvas(root, width=500, height=350)
canvas1.pack()

# PCM
label1 = tk.Label(root, text='                              PCM:( 540-790)')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry(root)
canvas1.create_window(270, 100, window=entry1)

# Sports
label2 = tk.Label(root, text='           Sports:(1.7-4)')
canvas1.create_window(120, 120, window=label2)

entry2 = tk.Entry(root)
canvas1.create_window(270, 120, window=entry2)

# Project
label3 = tk.Label(root, text='      Project_Stars:(1-6)')
canvas1.create_window(140, 140, window=label3)

entry3 = tk.Entry(root)
canvas1.create_window(270, 140, window=entry3)

# Warnings
label4 = tk.Label(root, text='Warnings:(22-31)')
canvas1.create_window(160, 160, window=label4)

entry4 = tk.Entry(root)
canvas1.create_window(270, 160, window=entry4)
```

```python
def values():
    global pcm_marks
    pcm_marks = float(entry1.get())

    global sport_grades
    sport_grades = float(entry2.get())

    global proj_marks
    proj_marks = float(entry3.get())

    global warnings
    warnings = float(entry4.get())

    y_pred = clf.predict([[pcm_marks, sport_grades, proj_marks, warnings]])
    Prediction_result = y_pred[0]
    print("Prediction_result is ", Prediction_result)
    if Prediction_result == 2:
        result = "Prediction : Will be admitted "
    elif Prediction_result==1:
        result = "Prediction : Will be in wait list "
    else:
        result = "Prediction : Will not be admitted "

    label_Prediction = tk.Label(root, text=result, bg='sky blue')
    canvas1.create_window(270, 280, window=label_Prediction,width=300)
```

```python
button1 = tk.Button(root, text='      Predict      ', command=values, bg='green', fg='white', font=11)
canvas1.create_window(270, 220, window=button1)
```

```python
root.mainloop()
```
